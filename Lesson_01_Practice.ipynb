```python
import math
import os

# 1. Create a file with some numbers (Simulating a data file)
with open("data.txt", "w") as f:
    f.write("16, 25, 36, 49")

# 2. Read the file
with open("data.txt", "r") as f:
    content = f.read()
    # Convert the string "16, 25..." into a list of actual numbers
    numbers = [int(n) for n in content.split(",")]

print(f"Reading numbers from file: {numbers}")

# 3. Use the Math library to find the Square Root of each number
print("\n--- Math Results ---")
for num in numbers:
    root = math.sqrt(num)
    print(f"The square root of {num} is {root}")
```

```python
import pandas as pd

# 1. Create some data (like a mini-spreadsheet)
data = {
    'Student': ['Raul', 'Gemini', 'Professor'],
    'Grade': [10, 9.5, 8.0],
    'Subject': ['AI', 'Python', 'History']
}

# 2. Convert to a "DataFrame" (This is how Python sees a table)
df = pd.DataFrame(data)

# 3. Save it to an Excel file
df.to_excel("grades_report.xlsx", index=False)
print("✅ Excel file 'grades_report.xlsx' has been created!")

# 4. Read it back to prove it works
df_from_excel = pd.read_excel("grades_report.xlsx")
print("\n--- Data read from the Excel file ---")
print(df_from_excel)
```

```python
import subprocess
import sys

def install_if_missing(package):
    try:
        __import__(package)
        print(f"✅ {package} is already installed in your .venv!")
    except ImportError:
        print(f"⚠️ {package} not found. Installing now...")
        # This command runs 'pip install' from inside your notebook
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"✅ {package} installed successfully!")

# List of libraries we want to check
libraries = ['pandas', 'openpyxl', 'matplotlib']

for lib in libraries:
    install_if_missing(lib)
```
