- Goal 1: Create a file.
- Goal 2: Read the file.

```python
# 1. Create a file called 'hello.txt'
with open("hello.txt", "w") as f:
    f.write("Environment check: Success!")

# 2. Read it back
with open("hello.txt", "r") as f:
    print(f.read())
```
