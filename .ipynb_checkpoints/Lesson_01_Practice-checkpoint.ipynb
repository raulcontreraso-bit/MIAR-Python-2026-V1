```python
# 1. Create a file called 'hello.txt'
with open("hello.txt", "w") as f:
    f.write("Environment check: Success!")

# 2. Read it back
with open("hello.txt", "r") as f:
    print(f.read())
    
```

```python
# Variables for time
days_per_week = 7
hours_per_day = 24
minutes_per_hour = 60
seconds_per_minute = 60

# The calculation
seconds_per_week = days_per_week * hours_per_day * minutes_per_hour * seconds_per_minute   

print(f"There are {seconds_per_week:,} seconds in a  week!")
```

```python
import math
import os

# 1. Create a file with some numbers (Simulating a data file)
with open("data.txt", "w") as f:
    f.write("16, 25, 36, 49")

# 2. Read the file
with open("data.txt", "r") as f:
    content = f.read()
    # Convert the string "16, 25..." into a list of actual numbers
    numbers = [int(n) for n in content.split(",")]

print(f"Reading numbers from file: {numbers}")

# 3. Use the Math library to find the Square Root of each number
print("\n--- Math Results ---")
for num in numbers:
    root = math.sqrt(num)
    print(f"The square root of {num} is {root}")
```
